```python
from abc import ABC, abstractmethod
import mysql.connector

class DML(ABC):
    def __init__(self):
        self.db = mysql.connector.connect(
            host="localhost",
            user="root",
            password="Mugi9678",
            database="project_P0"
        )
        self.cursor = self.db.cursor()

    @abstractmethod
    def insertion(self, data):
        pass

    def deletion(self):
        pass

    def updation(self):
        pass

    def __del__(self):
        # Close the cursor and connection when the object is destroyed
        try:
            if self.cursor:
                self.cursor.close()
            if self.db:
                self.db.close()
        except mysql.connector.Error as err:
            print(f"Error: {err}")

class CustomerInsertion(DML):
    def insertion(self, customer_data):
        insert_query = """
        INSERT INTO Customers (customer_id, customer_name, country, city)
        VALUES (%s, %s, %s, %s)
        """
        try:
            self.cursor.executemany(insert_query, customer_data)
            self.db.commit()
            print("Data inserted successfully.")
        except mysql.connector.Error as err:
            print(f"Error: {err}")
        finally:
            # You can also choose to close the connection in the `__del__` method
            self.cursor.close()
            self.db.close()
```

```python
# Create an instance of CustomerInsertion
data = CustomerInsertion()

# Data to insert
customer_data = [(1, "naga", "India", "mysuru")]

# Insert data into the Customers table
data.insertion(customer_data)

# Update data
new_data = ("John", "USA", "New York")
data.updation(1, new_data)
```

```python
import pandas as pd
import mysql.connector
import time

MAX_RETRIES = 5
RETRY_DELAY = 2 

# Load JSON data
customer_df = pd.read_json('customers.json')  # Ensure this path is correct
transaction_logs_df = pd.read_json('transaction_logs.json')  # Ensure this path is correct


transaction_logs_df['datetime'] = pd.to_datetime(transaction_logs_df['datetime'])
transaction_logs_df['payment_txn_success'] = transaction_logs_df['payment_txn_success'].astype(bool)

for attempt in range(MAX_RETRIES):
    conn = None
    cursor = None
    try:
        # Connect to MySQL
        conn = mysql.connector.connect(
            host='localhost',
            user='root',  
            password='Mugi9678',  
            database='projectp0'
        )
        cursor = conn.cursor()

        # Insert data into Customers table
        for index, row in customer_df.iterrows():
            cursor.execute("""
                INSERT INTO Customers (customer_id, customer_name, country, city)
                VALUES (%s, %s, %s, %s)
                ON DUPLICATE KEY UPDATE
                customer_name = VALUES(customer_name),
                country = VALUES(country),
                city = VALUES(city)
            """, (row['customer_id'], row['customer_name'], row['country'], row['city']))

        # Insert data into Products table
        products_df = transaction_logs_df[['product_id', 'product_name', 'product_category']].drop_duplicates()

        for index, row in products_df.iterrows():
            cursor.execute("""
                INSERT INTO Products (product_id, product_name, product_category)
                VALUES (%s, %s, %s)
                ON DUPLICATE KEY UPDATE
                product_name = VALUES(product_name),
                product_category = VALUES(product_category)
            """, (row['product_id'], row['product_name'], row['product_category']))

        # Insert data into Orders table
        for index, row in transaction_logs_df.iterrows():
            cursor.execute("""
                INSERT INTO Orders (order_id, customer_id, product_id, qty, price, datetime, ecommerce_website_name)
                VALUES (%s, %s, %s, %s, %s, %s, %s)
            """, (row['order_id'], row['customer_id'], row['product_id'], row['qty'], row['price'], row['datetime'], row['ecommerce_website_name']))

        # Insert data into Payments table
        for index, row in transaction_logs_df.iterrows():
            cursor.execute("""
                INSERT INTO Payments (order_id, payment_type, payment_txn_id, payment_txn_success, failure_reason)
                VALUES (%s, %s, %s, %s, %s)
            """, (row['order_id'], row['payment_type'], row['payment_txn_id'], row['payment_txn_success'], row['failure_reason']))

        # Commit the transaction
        conn.commit()
        print("Data inserted successfully.")
        break  # Exit loop if successful

    except mysql.connector.Error as err:
        print(f"Error: {err}")
        if err.errno == 1045:  # Access denied
            print("Access denied for user. Please check credentials.")
            break  # Stop retrying
        elif err.errno == 1205:  # Lock wait timeout
            print(f"Lock wait timeout exceeded. Retrying ({attempt + 1}/{MAX_RETRIES})...")
        elif err.errno == 2006:  # MySQL server has gone away
            print("MySQL server has gone away. Reconnecting...")
        elif err.errno == 2055:  # Cursor is not connected
            print("Cursor is not connected. Reconnecting...")
        time.sleep(RETRY_DELAY)
        if conn and conn.is_connected():
            conn.rollback()  # Rollback the transaction
            cursor.close()
            conn.close()
    except Exception as e:
        print(f"Unexpected error: {e}")
        if conn and conn.is_connected():
            conn.rollback()
            cursor.close()
            conn.close()
        break
    finally:
        # Ensure resources are cleaned up
        if conn and conn.is_connected():
            cursor.close()
            conn.close()
```

```python
import mysql.connector
import pandas as pd

# Define the execute_query function
def execute_query(query):
    try:
        # Connect to MySQL
        conn = mysql.connector.connect(
            host='localhost',
            user='root',
            password='Mugi9678',
            database='projectp0'
        )

        # Create a cursor and execute the query
        cursor = conn.cursor()
        cursor.execute(query)

        # Fetch the results and convert them to a DataFrame
        result = cursor.fetchall()
        columns = [col[0] for col in cursor.description]  # Get column names from cursor
        df = pd.DataFrame(result, columns=columns)

        return df

    except mysql.connector.Error as err:
        print(f"Error: {err}")
        return None

    finally:
        # Ensure resources are cleaned up
        if conn.is_connected():
            cursor.close()
            conn.close()

# Execute the query and print the record count for the Customers table
query = "SELECT COUNT(*) AS count FROM Customers"
df = execute_query(query)
if df is not None:
    actual_count = df.iloc[0]['count']
    print(f"Customers table record count: {actual_count}")

# Execute the query and print the record count for the Products table
query = "SELECT COUNT(*) AS count FROM Products"
df = execute_query(query)
if df is not None:
    actual_count = df.iloc[0]['count']
    print(f"Products table record count: {actual_count}")

# Testing Orders Table Record Count
query = "SELECT COUNT(*) AS count FROM Orders"
df = execute_query(query)
if df is not None:
    actual_count = df.iloc[0]['count']
    print(f"Orders table record count: {actual_count}")

# Testing Payments Table Record Count
query = "SELECT COUNT(*) AS count FROM Payments"
df = execute_query(query)
if df is not None:
    actual_count = df.iloc[0]['count']
    print(f"Payments table record count: {actual_count}")
```

```python
query = """
SELECT * FROM Orders o
LEFT JOIN Customers c ON o.customer_id = c.customer_id
WHERE c.customer_id IS NULL
"""
df = execute_query(query)
if df is not None:
    if df.empty:
        print("Foreign key constraints are correct.")
    else:
        print("Foreign key constraints issue found:")
        print(df)

query = """
SELECT * FROM Orders o
LEFT JOIN Products p ON o.product_id = p.product_id
WHERE p.product_id IS NULL
"""
df = execute_query(query)
if df is not None:
    if df.empty:
        print("Foreign key constraints are correct.")
    else:
        print("Foreign key constraints issue found:")
        print(df)
```

```python
query = """
SELECT product_id, COUNT(*) as count
FROM Products
GROUP BY product_id
HAVING count > 1
"""
df = execute_query(query)
if df is not None:
    if df.empty:
        print("No duplicate records found in Products table.")
    else:
        print("Duplicate records found in Products table:")
        print(df)

query = """
SELECT order_id, COUNT(*) as count
FROM Orders
GROUP BY order_id
HAVING count > 1
"""
df = execute_query(query)
if df is not None:
    if df.empty:
        print("No duplicate records found in Orders table.")
    else:
        print("Duplicate records found in Orders table:")
        print(df)
```
